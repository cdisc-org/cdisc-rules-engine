```python
import pandas as pd
import json
import pickle

```

```python
with open("rules.json", "r", encoding="utf-8") as f:
    rules_data = json.load(f)

df = pd.DataFrame(rules_data)
df.head(5)
```

```python

with open("../../../resources/cache/rules.pkl", "rb") as f:
    rules = pickle.load(f)
rules = pd.DataFrame(rules).transpose()
rules.head(5)
```

```python

test_data = df.explode("positive_regressions")
test_data = test_data[test_data["positive_regressions"].notnull()]
test_data['positive_regressions'] = test_data['positive_regressions'].apply(lambda v: next(val['engine_regression'] for val in v.values()))
test_data = test_data[test_data['positive_regressions'].notnull()]
test_data.head(5)
```

```python
test_data['comparison'] = test_data["positive_regressions"].map(lambda v: v.get('old_vs_sql', None))
test_data = test_data[test_data['comparison'].notnull()]
test_data = test_data[[c.get("equal", None) is None for c in test_data['comparison']]]
test_data = test_data[[c.get("number_of_errors_mismatch", False) for c in test_data['comparison']]]
test_data[10:15]
```

```python
simple_rule_types = [
    'Record Data',
    'Dataset Contents Check against Define XML and Library Metadata',
    'Value Level Metadata Check against Define XML'
]
not_simple_rules = rules[[r not in simple_rule_types for r in rules['rule_type']]]
print(len(not_simple_rules), "not simple rules")
print(not_simple_rules['rule_type'].unique())
not_simple_rules['core_id'].sort_values().head(5)
```
