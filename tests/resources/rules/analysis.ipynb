```python
import pandas as pd
import json
import pickle

```

```python
with open("rules.json", "r", encoding="utf-8") as f:
    rules_data = json.load(f)

df = pd.DataFrame(rules_data)
df.head(5)
```

```python

with open("../../../resources/cache/rules.pkl", "rb") as f:
    rules = pickle.load(f)
rules = pd.DataFrame(rules).transpose()
rules.head(5)
```

```python

pos_test_data = df.explode("positive_regressions")
pos_test_data = pos_test_data[pos_test_data["positive_regressions"].notnull()]
pos_test_data['positive_regressions'] = pos_test_data['positive_regressions'].apply(lambda v: next(val['engine_regression'] for val in v.values()))
pos_test_data = pos_test_data[pos_test_data['positive_regressions'].notnull()]
pos_test_data['comparison'] = pos_test_data["positive_regressions"].map(lambda v: v.get('old_vs_sql', None))
pos_test_data.head(5)

neg_test_data = df.explode("negative_regressions")
neg_test_data = neg_test_data[neg_test_data["negative_regressions"].notnull()]
neg_test_data['negative_regressions'] = neg_test_data['negative_regressions'].apply(lambda v: next(val['engine_regression'] for val in v.values()))
neg_test_data = neg_test_data[neg_test_data['negative_regressions'].notnull()]
neg_test_data['comparison'] = neg_test_data["negative_regressions"].map(lambda v: v.get('old_vs_sql', None))
neg_test_data.head(5)
```

```python
simple_rule_types = [
    'Record Data',
    'Dataset Contents Check against Define XML and Library Metadata',
    'Value Level Metadata Check against Define XML'
]
not_simple_rules = rules[[r not in simple_rule_types for r in rules['rule_type']]]
not_simple_rule_ids = not_simple_rules['core_id'].unique()
simple_rules = rules[[r in simple_rule_types for r in rules['rule_type']]]
simple_rule_ids = simple_rules['core_id'].unique()
print(len(not_simple_rules), "not simple rules")
print(not_simple_rules['rule_type'].unique())
not_simple_rules['core_id'].sort_values().head(5)
```

```python
errors = neg_test_data[neg_test_data['comparison'].notnull()]
errors = errors[[c in simple_rule_ids for c in errors['core-id']]]
errors = errors[[c.get("equal", None) is None for c in errors['comparison']]]
errors = errors[[c.get("diff", False) != {} for c in errors['comparison']]]
print(len(errors))
errors.head(10)
```
